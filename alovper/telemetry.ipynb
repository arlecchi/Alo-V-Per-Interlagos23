```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
# Setup plotting
ff1.plotting.setup_mpl()
```

```python
race = ff1.get_session(2023, 'Interlagos', 'R')
race.load()
```

```python
laps_per = race.laps.pick_driver('PER')
laps_alo = race.laps.pick_driver('ALO')
lapn = 71
```

```python
per_lap = laps_per[laps_per['LapNumber'] == 71].iloc[0]
alo_lap = laps_alo[laps_alo['LapNumber'] == 71].iloc[0]
```

```python
per_tel = per_lap.get_car_data().add_distance().add_driver_ahead()
alo_tel = alo_lap.get_car_data().add_distance().add_driver_ahead()
```

```python
per_color = ff1.plotting.driver_color('PER')
alo_color = ff1.plotting.driver_color('ALO')
```

```python
fig, ax = plt.subplots(7)
fig.suptitle("Lap 71 Battle Telemetry Comparison Based on Time")

ax[0].plot(per_tel['Time'], per_tel['Speed'], label='PER', color = per_color)
ax[0].plot(alo_tel['Time'], alo_tel['Speed'], label='ALO', color = alo_color)
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(per_tel['Time'], per_tel['Throttle'], label='PER', color = per_color)
ax[1].plot(alo_tel['Time'], alo_tel['Throttle'], label='ALO', color = alo_color)
ax[1].set(ylabel='Throttle')

ax[2].plot(per_tel['Time'], per_tel['Brake'], label='PER', color = per_color)
ax[2].plot(alo_tel['Time'], alo_tel['Brake'], label='ALO', color = alo_color)
ax[2].set(ylabel='Brakes')

ax[3].plot(per_tel['Time'], per_tel['RPM'], label='PER', color = per_color)
ax[3].plot(alo_tel['Time'], alo_tel['RPM'], label='ALO', color = alo_color)
ax[3].set(ylabel='RPM')

ax[4].plot(per_tel['Time'], per_tel['nGear'], label='PER', color = per_color)
ax[4].plot(alo_tel['Time'], alo_tel['nGear'], label='ALO', color = alo_color)
ax[4].set(ylabel='Gear')

ax[5].plot(per_tel['Time'], per_tel['DRS'], label='PER', color = per_color)
ax[5].plot(alo_tel['Time'], alo_tel['DRS'], label='ALO', color = alo_color)
ax[5].set(ylabel='DRS')

ax[6].plot(per_tel['Time'], per_tel['DistanceToDriverAhead'], label='PER', color = per_color)
ax[6].plot(alo_tel['Time'], alo_tel['DistanceToDriverAhead'], label='ALO', color = alo_color)
ax[6].set(ylabel='Dist')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig, ax = plt.subplots(7)
fig.suptitle("Lap 71 Battle Telemetry Comparison Based on Distance")

ax[0].plot(per_tel['Distance'], per_tel['Speed'], label='PER', color = per_color)
ax[0].plot(alo_tel['Distance'], alo_tel['Speed'], label='ALO', color = alo_color)
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(per_tel['Distance'], per_tel['Throttle'], label='PER', color = per_color)
ax[1].plot(alo_tel['Distance'], alo_tel['Throttle'], label='ALO', color = alo_color)
ax[1].set(ylabel='Throttle')

ax[2].plot(per_tel['Distance'], per_tel['Brake'], label='PER', color = per_color)
ax[2].plot(alo_tel['Distance'], alo_tel['Brake'], label='ALO', color = alo_color)
ax[2].set(ylabel='Brakes')

ax[3].plot(per_tel['Distance'], per_tel['RPM'], label='PER', color = per_color)
ax[3].plot(alo_tel['Distance'], alo_tel['RPM'], label='ALO', color = alo_color)
ax[3].set(ylabel='RPM')

ax[4].plot(per_tel['Distance'], per_tel['nGear'], label='PER', color = per_color)
ax[4].plot(alo_tel['Time'], alo_tel['nGear'], label='ALO', color = alo_color)
ax[4].set(ylabel='Gear')

ax[5].plot(per_tel['Distance'], per_tel['DRS'], label='PER', color = per_color)
ax[5].plot(alo_tel['Distance'], alo_tel['DRS'], label='ALO', color = alo_color)
ax[5].set(ylabel='DRS')

ax[6].plot(per_tel['Distance'], per_tel['DistanceToDriverAhead'], label='PER', color = per_color)
ax[6].plot(alo_tel['Distance'], alo_tel['DistanceToDriverAhead'], label='ALO', color = alo_color)
ax[6].set(ylabel='Dist')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for a in ax.flat:
    a.label_outer()
    
plt.show()
```
